```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_org = pd.read_csv('result.csv')

widget_id_list = ['6Xg9xW7N', 'y74kVNoV']
df_list = []
```

```python
for widget_id in widget_id_list:
    df_splitted = df_org.groupby('widget_id').get_group(widget_id)
    df_list.append(df_splitted)
```

```python
def draw_scatter(df, widget_id):
    df = df[(df['click_x_group'] < 330) & (df['click_y_group'] < 55)]
    max_cnt = df['click_cnt'].max()
    plt.title(widget_id)
    plt.scatter(df['click_x_group'], df['click_y_group'], s=df['click_cnt']/max_cnt*100)
    plt.savefig(fname=widget_id+'_scatter.pdf', bbox_inches='tight', pad_inches=0.1)
    plt.show()
```

```python
for i in range(len(df_list)):
    widget_id = widget_id_list[i]
    df = df_list[i]
    draw_scatter(df, widget_id)
```
